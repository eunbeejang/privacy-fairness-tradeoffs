```python
import numpy as np
import matplotlib.pyplot as plt
```

# Graphing initial results for BANK&ADULT datasets

## 1. Bank Marketing Dataset

```python
        # acc, recall, eq_odds, dem_par
bank = [[91.74, 61.19, 70.46, 13.57], # original
        [83.11, 89.43, 94.12, 31.73], # dp_sgd 0.1
        [81.40, 90.46, 96.65, 36.12], # dp_sgd 1.0
        [84.29, 8.20, 13.48, 13.17],  # rand 0.1
        [82.86, 11.09, 15.38, 11.16], # ind 0.1
        [84.68, 5.1, 10.43, 10.09],   # corr 0.1
        [85.55, 5.76, 17.65, 12.57],  # rand 1.0
        [79.53, 9.98, 20.93, 13.94],  # ind 1.0
        [83.37, 7.76, 13.85, 10.28]]  # corr 1.0

bar_width = 0.05    
br1 = np.arange(4) 

br2 = [x + bar_width*2 for x in br1]
br3 = [x + bar_width for x in br2]

br4 = [x + bar_width*2 for x in br3]
br5 = [x + bar_width for x in br4]
br6 = [x + bar_width for x in br5]

br7 = [x + bar_width*2 for x in br6]
br8 = [x + bar_width for x in br7]
br9 = [x + bar_width for x in br8]



x_label = ['Accuracy', 'Recall', 'Equalized Odds', 'Demographic Parity']
legend = ['original', 'dp_sgd_0.1', 'dp_sgd_1.0',
          'rand_0.1', 'ind_0.1', 'corr_0.1',
          'rand_1.0', 'ind_1.0', 'corr_1.0']


fig = plt.figure(figsize =(10, 5))
ax = fig.add_axes([0,0,1,1])
ax.bar(br1, bank[0], color = 'b', width = bar_width)

ax.bar(br2, bank[1], color = 'g', width = bar_width)
ax.bar(br3, bank[2], color = 'c', width = bar_width)

ax.bar(br4, bank[3], color = 'orange', width = bar_width)
ax.bar(br5, bank[4], color = 'tomato', width = bar_width)
ax.bar(br6, bank[5], color = 'purple', width = bar_width)

ax.bar(br7, bank[6], color = 'dodgerblue', width = bar_width)
ax.bar(br8, bank[7], color = 'peru', width = bar_width)
ax.bar(br9, bank[8], color = 'navy', width = bar_width)
ax.legend(legend,loc=1)
plt.title("Bank Marketing Dataset (Sensitive Variable: Education)")

plt.xticks([r + bar_width*4 for r in range(4)], x_label) 
plt.show()
```
